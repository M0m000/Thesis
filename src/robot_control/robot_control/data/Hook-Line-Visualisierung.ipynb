```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def plot_3d_lines_between_points(point_pairs):
    """
    Zeichnet 3D-Gerade zwischen mehreren Punktpaaren in einem breiten 20x5-Plot.
    
    Parameters:
    - point_pairs: Liste von Tupeln [(P1, P2), (P3, P4), ...]
      wobei P1, P2 usw. Listen oder Tupel der Form [x, y, z] sind.
    """
    fig = plt.figure(figsize=(10, 10))
    ax = fig.add_subplot(111, projection='3d')

    for i, (P1, P2) in enumerate(point_pairs):
        X = [P1[0], P2[0]]
        Y = [P1[1], P2[1]]
        Z = [P1[2], P2[2]]

        ax.plot(X, Y, Z, linewidth=2, label=f'Linie {i+1}')
        ax.scatter(*P1, color='red', s=30)
        ax.scatter(*P2, color='green', s=30)

    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title('3D-Gerade zwischen mehreren Punktpaaren')
    ax.set_box_aspect([1,1,1])
    ax.auto_scale_xyz(ax.get_xlim(), ax.get_ylim(), ax.get_zlim())
    ax.legend()
    plt.grid(True)
    plt.show()
```

```python
lines = [
    ([0, 0, 0], [1, 1, 1]),
    ([1, 2, 3], [4, 5, 6]),
    ([2, 0, 1], [2, 3, 5])
]

plot_3d_lines_between_points(lines)
```
